**Data Cleaning** 

```python
#Removing  Unnecessary Column 
```

```python
df.drop('customerID' ,axis=1 ,inplace =True)
```

```python
df.info()
```

Fixing 'TotalCharges' datatype  --- its is given as object but it should be float64

```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df.info()
```

```python
df['TotalCharges'].isna().sum()
```

```python
df = df.dropna(subset=['TotalCharges'])
```

```python
df['TotalCharges'].isna().sum()
```

```python
df.isna().sum()
```

**We are replacing 'No Internet services' and 'No phone services' with 'NO'**  ....            
Because for business  both mean service not used             
From modeling/visualization perspective, both behave the same.

```python
df.replace('No internet service' , 'No' ,inplace= True)
df.replace('No phone service' ,'No', inplace =True)
df.head(5)
```

**Convert Yes / No → 0 / 1---> Binary Encoding**              
WE encode only true binary variables used for calculations while keeping categorical service features as text to preserve interpretability and improve dashboard visualization clarity. 

We convert to 0/1 ONLY when the column is truly binary and represents a single decision or state.
Meaning: only two meaningful outcomes exist.


```python
binary_cols = [
    'Partner','Dependents','PhoneService',
    'PaperlessBilling','Churn'
]

for col in binary_cols:
    df[col] = df[col].map({'Yes':1,'No':0})
```

```python
df.head()
```

Convert SeniorCitizen Data type   from Int64 to category 

```python
df['SeniorCitizen'] = df['SeniorCitizen'].astype('category')
```

```python
df.info()
```

**FEATURE ENGINEERING**  (Making New Columns)

Tenure Group

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0,12,24,48,72],
    labels=['0-1yr','1-2yr','2-4yr','4-6yr'],
    include_lowest=True
)
```

```python
df.head()
```

Contract Encoding          
Contract types represent customer commitment level

```python
df['ContractEncoded'] = df['Contract'].map({
    'Month-to-month':0,
    'One year':1,
    'Two year':2
})
```

Total Services Count         
How many telecom services each customer is using.     
Instead of looking at 6 separate columns, we create one summary number.       

```python
services = [
'OnlineSecurity','OnlineBackup','DeviceProtection',
'TechSupport','StreamingTV','StreamingMovies'
]

df['TotalServices'] = df[services].apply(
    lambda x: (x=='Yes').sum(), axis=1)
```

Average Monthly Spend    
On average, how much money a customer spends per month.

```python
df['AvgMonthlySpend'] = df['TotalCharges'] / df['tenure']
df['AvgMonthlySpend'] = df['AvgMonthlySpend'].fillna(df['MonthlyCharges'])
```
